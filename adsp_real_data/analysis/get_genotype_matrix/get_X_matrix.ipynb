1. Change cutoff for percentage of NAs in CNV-level from 0.4 to 0.8, which means I include more CNVs.

```python
import pandas as pd
import numpy as np
import os
cwd = os.path.expanduser("~/cnv-project/adsp_real_data")
ref_gene = pd.read_table(f"{cwd}/refGene.clean.gz", compression = "gzip", sep = "\t", header = 0)
```

```python
del_comm_cnv = pd.read_csv(f"{cwd}/data/del_comm_adsp_hq.gz", compression="gzip", sep="\t")
del_comm_cnv = del_comm_cnv.drop(columns=["COUNTED","ALT"])
del_comm_cnv.iloc[:,0] = del_comm_cnv.iloc[:,0].str.replace('chr','')

n_cnv,n_cols = del_comm_cnv.shape

percent_na_indv = [del_comm_cnv.iloc[:,i].isna().sum() / n_cnv for i in range(4,n_cols)]
cutoff_indv = 0.2
indv_to_keep = [index for index, element in enumerate(percent_na_indv) if element <= cutoff_indv]
del_comm_cnv = del_comm_cnv.iloc[:,indv_to_keep]

n_indvs = del_comm_cnv.shape[1] - 4
percent_na_cnv = [del_comm_cnv.iloc[i,:].isna().sum() / (n_indvs) for i in range(n_cnv)] 
cutoff_cnv = 0.8
cnv_to_keep = [index for index, element in enumerate(percent_na_cnv) if element <= cutoff_cnv]
del_comm_cnv = del_comm_cnv.iloc[cnv_to_keep,:]
```

```python
cnv_in_chr22 = del_comm_cnv[del_comm_cnv['CHR'] == '22'].iloc[:,range(4)]
genes_in_chr22 = ref_gene[ref_gene['CHR'] == '22']
```

```python
cnv_in_chr22
```

```python
genes_in_chr22
```

```python
chr22_df = cnv_in_chr22.merge(genes_in_chr22, how='left', on='CHR')
chr22_df
```

```python
mask = (
        (chr22_df['CM'] >= chr22_df['start']) & (chr22_df['CM'] <= chr22_df['end']) |
        (chr22_df['CM'] <= chr22_df['start']) & (chr22_df['POS'] >= chr22_df['end']) |
        (chr22_df['POS'] >= chr22_df['start']) & (chr22_df['POS'] <= chr22_df['end'])
       )
overlap_df = chr22_df[mask]
overlap_df
```

```python
overlap_df_indv = overlap_df.merge(del_comm_cnv[['SNP','0:A-ACT-AC000014-BL-NCR-15AD78694']], how='inner', on='SNP')
overlap_df_indv
```

```python
gene_dict = {}
for g in overlap_df_indv['gene'].unique():
    gene_dict[g] = []
    if (overlap_df_indv[overlap_df_indv['gene'] == g]['0:A-ACT-AC000014-BL-NCR-15AD78694']).sum() != 0:
        gene_dict[g].append(1)
    else:
        gene_dict[g].append(0)
        
res = pd.DataFrame(gene_dict)
res
```
