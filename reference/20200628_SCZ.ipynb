# Common schizophrenia alleles are enriched in mutation-intolerant genes and in regions under strong background selection
Antonio F. Pardinas, et al

## Introduction
- Schizophrenia (SCZ) is highly heritable and polygenic
    - variants span the full spectrum of population frequencies
    - common alleles explain between 1/3 and 1/2 of the genetic variance inliability


- Various hypotheses have been proposed about how common risk allelespersist in the population:
    - compensatory advantage (balancing selection)
    - hitchhiking
    - contrasting theories

The authors report a new genome-wide association study (GWAS) of SCZ and meta-analysis with existing data to identify existing and novel disease associated loci.

## Data
Meta-analysis data sources:
- Psychiatric Genomics Consortium (PGC)
- Genome-wide genotype information for SCZ cases from the UK (CLOZUK)
- Public repositories
- Collaboration

Sample size:
- Meta-analysis: 40,675 cases and 64,643 controls
- CLOZUK etc: 11,260 cases and 24,542 controls
- New samples: 5,220 cases and 18,823 controls

## Methods
### CLOZUK SCZ GWAS (11,260 cases, 24,542 controls)
- Logistic regression
- Adjusted for 11 PCA covariates
- Performed regular LD clumping to obtain independent index SNPs

### Meta-analysis (40,675 cases and 64,643 controls)
- Independent PGC samples (29,415 cases and 40,101 controls)
- CLOZUK

### Identify mutation-intolerant genes
#### Gene set analysis
- MAGMA v1.03 on meta-analysis summary statistics
- Exclude xMHC region and X chromosome
- Calculate gene-wide Pvalue by combining Pvalues of all SNPs inside a gene
- Allow for 35 kb upstream and 10 kb downstream
- Calculate gene set Pvalues on the gene-wide Pvalues

#### Partitioned linkage disequilibrium score regression (LDSR)
- Finucane HK, et al. Nat Genet. 2015
- To quantify the contribution of SNPs within LoF-intolerant genes to SCZ SNP-based heritability

### Systems genomics
- Use MAGMA
- Perform a primary analysis of 134 central nervous system (CNS)-related gene sets which capture the excess copy number variation (CNV) burden in SCZ

### Common risk alleles in regions under background selection
Aim to explore whether some form of natural selection is linked to the maintenance of common genetic risk in SCZ.
- 4 genome-wide metric of positive selection, one background selection, one Neanderthal introgression
- All SNPs were assigned a value for each measure (directly or linear interpolation)
- Binary annotations: dichotomize at extreme cutoffs of top 2%, 1% and 0.5% of the values in the full set of SNPs
- Derived $τ_c$ coefficient of enriched natural selection annotations

### Identifying likely candidate SNPs
- Use FINEMAP to highlight credibly causal variants at each of the 145 identified loci.
- Variants were annotated with ANNOVAR.
- Mapped SNPs with posterior probability higher than 0.5.

## Results
### GWAS and meta-analysis
- Identified 179 independent genome-wide significant SNPs ($P<5×10^{−8}$) mapping to 145 independent loci.
- 145 loci included 93 significant loci in PGC study.

### Gene set analysis for mutation-intolerant genes
- Loss-of-function (LoF) intolerant genes were enriched for common variant associations with SCZ in comparison with all other annotated genes.
- Alzheimer’s disease, type 2 diabetes and neuroticism (late-onset) showed weak signals for enrichment of the LoF-intolerant gene set.

### LDSR
- Overall, genic SNPs accounted for 64% of $h^2_{SNP}$ (1.23-foldenrichment).
- LoF-intolerant genes explained 30% of $h^2_{SNP}$ (47% of all genic $h^2_{SNP}$, 2.01-fold enrichment).
- Genes classed as not LoF intolerant were significantly depleted for $h^2_{SNP}$ (0.90-fold).

### System Genomics
- Capture a high fraction of $h^2_{SNP}$ (30% total, 46% genic).
- 54 out of 134 gene sets were nominally significant.
- The most strongly associated gene set constituted the targets of the fragile X mental retardation protein (FMRP).
- Associated gene sets also include voltage-gated calcium channel complexes and the $\text{5-HT}_{2C}$ receptor complex.

### Common risk alleles in regions under background selection
- Observed strong evidence for SCZ $h^2_{SNP}$ enrichment in SNPs under strong background selection (BGS).
- Found a significant depletion of $h^2_{SNP}$ in SNPs subject to positive selection.
- Found no significant relationship between $h^2_{SNP}$ and other positive selection or Neanderthal introgression meassures.
