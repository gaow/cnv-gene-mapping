## Enrichment analyses
### Single

We first examined the univariate enrichment of SCZ risk variants in ASoC SNPs, OCRs and other annotated genomic regions, by using TORUS. Red dots and black bars represent enrichment odds ratio and 95% confidence interval, respectively. If the confidence interval does not include 0, we call the enrichment result statistically significant.

iN-Glut ASoC variants showed over 100-fold remarkable enrichment for SCZ risk variants, while NPC ASoC variants showed ~66-fold enrichment. In contrast, both iN-Glut and NPC OCRs revealed only ~3-fold enrichments (Fig 1). In other words, the function of neuronal ASoC iN-Glut and NPC variants are more highlighted than that in their corresponding OCRs. Other OCRs (iN-GA, iN-DN, iPSC) indicated significant enrichment as well. Nevertheless, the bar widths for ASoC variants are greater than those of OCRs, for the reason that the number of SCZ variants that overlap with ASoC SNPs is much fewer than those captured by OCR peaks.

It has been widely proven that SCZ risk loci are enriched in non-coding regulatory regions (1). We verified that conserved and 3'UTR sequences were slightly more enriched for SCZ risk variants than coding regions. Furthermore, SCZ risk variants were also highly enriched in 5'UTR and promoter sequences. We further performed single-annotation analysis using brain eQTL for comparison, since it has been demonstrated that aggregate associations between brain eQTLs and many brain disorder GWAS variants, such as SCZ (2). We discovered that brain eQTL revealed ~4-fold enrichment for SCZ risk variants (Fig 1). Specifically, it should be noted that repressed regions, also called inactive regions (3), showed negative enrichment (odds ratio less than 1) of SCZ risk variants, because it is strongly indicated that the genes within repressed regions are silenced (3).

### Scatter plot

We merged fine-mapping results of SNPs from all 108 GWAS loci and then compared the effects of different functional genomic features more directly in one figure (Fig. 12). The x-axis represents PIPs from uniform priors, while y-axis represents PIPs from joint ASoC and OCRs priors. Red dots represent GWAS SNPs overlapped with either of 2 types of ASoC SNPs; yellow dots represent GWAS SNPs captured by OCRs; blue dots represent GWAS SNPs not located within OCRs. It is easily seen that most blue dots locate on or below the diagonal, indicating that SNPs not associated with ASoC or OCRs tend to have equal or lower PIPs when integrating functional feature priors for fine-mapping. In contrast, most red and yellow dots locate above the diagonal, specifically the red dots showed strong difference between two priors, highlighting ASoC as a functional mechanism of non-coding neuropsychiatric risk variants. This integrated method of fine-mapping with functional genomics data provides a powerful framework for identifying disease functional variants/genes.

Figure 12. Directly comparison of SuSiE PIPs from different priors for 108 SCZ GWAS significant loci.

Our fine-mapping analysis using SuSiE reported PIPs of SCZ variants. To estimate the percentage contribution of ASoC SNPs to disease risk, we summed over the PIPs of all ASoC iN-Glut and NPC SNPs, and divided that by the total PIPs of all SCZ GWAS SNPs. Consequently, we detected that non-coding SNPs in OCRs accounted for 9.4% of SCZ genome-wide SNPs and explained 22.5% of SCZ risk. In contrast, 2 types of ASoC SNPs, which accounted for only 0.08% of the SCZ SNPs, yet explained 5.5% of SCZ risk, a 67-fold enrichment. Considering that ASoC SNPs are only a small proportion of SNPs located in OCRs, ASoC SNPs are more enriched for SCZ GWAS risk variants, which justifies our proposition of utilizing ASoC to prioritize those risk variants that are more likely to be functional and potentially causal. This enrichment and fine-mapping approach is equally applicable to other neurodevelopmental brain disorders.

## VPS45

Of the top 20 ASoC SNPs, rs2027349 (chr1:150,039,678, GRCh37/hg19. All genomic positions are based on hg19) showed the strongest ASoC in NPC (Table S1) and second strongest ASoC in Glut (Table S2). rs2027349 locates at the VPS45 gene (chr1:150,039,342-150,117,505) that potentially relevant to SCZ given its involvement in neurotransmitter release (1). We explored the SCZ-associated genetic locus (chr1:149,998,890-150,242,490) where VPS45 gene resides. rs2027349 is the only local SNP with a high PIP (0.477, Fig. 15), while PIPs for all the other SNPs are constrained to almost 0. rs2027349, reference and alternative alleles are A and G, locates in a transcription factor (TF)-binding footprint, with the G allele disrupts the EGR3 binding motif, a TF involved in neurodevelopment and SCZ (2). Therefore, rs2027349 is prioritized for CRISPR/Cas9 editing. Our collaborators implemented CRISPR/Cas9 editing for ASoC SNP rs2027349 at VPS45 gene, and identified it as a highly likely SCZ risk SNP that affects VPS45 expression.
